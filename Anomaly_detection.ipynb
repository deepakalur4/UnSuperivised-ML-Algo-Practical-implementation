```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sen
```

```python
sen.displot(np.random.normal(loc=50, scale=5, size=100),kde=True)
```

```python
data=pd.DataFrame(np.random.normal(loc=50, scale=5, size=100),columns=["Data"])
```

```python
data.loc[5]=200
data.loc[10]=180
data.loc[15]=150
```

```python
plt.scatter(data.index,data["Data"])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
model=IsolationForest(n_estimators=20,contamination=0.1)
```

```python
model.fit_predict(data[["Data"]])
```

```python
plt.scatter(data.index,data["Data"],c=model.fit_predict(data[["Data"]]))
```

# Local outlier Factor

```python
x=np.random.randint(100,200,100)
```

```python
data=pd.DataFrame(x,columns=["Data"])
```

```python
data.loc[5]=400
data.loc[15]=300
data.loc[55]=500
```

```python
plt.scatter(data.index,data["Data"])
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
model=LocalOutlierFactor(n_neighbors=20)
```

```python
model.fit_predict(data[["Data"]])
```

```python
plt.scatter(data.index,data["Data"],c=model.fit_predict(data[["Data"]]))
```
